```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. 畫出 cos 圖檔，並儲存
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.title('cos')
```

```python
#2. 給定散點圖顏色
X = np.random.normal(0,100,100)
Y = np.random.normal(0,100,100)
plt.scatter(X,Y, s=75, c='r', alpha=.5)
plt.show()
```
